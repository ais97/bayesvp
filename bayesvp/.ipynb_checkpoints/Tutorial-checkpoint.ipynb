# BayesVP Tutorial

**Step 1: Download BayesVP**

Enter the following in the terminal in the directory you wish to install BayesVP

    git clone https://github.com/ais97/bayesvp.git
    
**Step 2: Install BayesVP to use from any directory**

Run setup.py
    
    python3 /path/to/bayesvp/setup.py develop 
    
**OR**
    
You may have to use the sudo command and use only "python"

    sudo python /path/to/bayesvp/setup.py develop

Now you are all set to use BayesVP!

**Step 3: Test if the codes are working or jump directly into your spectrum**

You can remove the "example" folder and save it in another directory.
(The best way is not to have the bayesvp path and your output paths as same)

Open bayesvp_try.py with your favourite Python editor and run all the imports.

1. To make new config files, run only lines 24 and 25. This should ask you for inputs like this:
        
        >>>config = wc.WriteBayesVPConfig()
        >>>config.print_to_file(interactive_write=(True))

        filename for output chain: o6

        state: VI

        Ending wavelength: 1033
        Now enter the priors. Press Enter for default values.
        min logN = 10
        max logN = 18
        min b = 0
        max b = 100
        central redshift = 0
        velocity range [km/s] = 100
        Now enter the MCMC parameters..
        Number of walkers: 200

        MCMC sampler kombine(default), emcee:

        Written config file: /home/aiswarya/bayesvp/bayesvp/data/example/config_OVI.dat
    
    Check the output config file to see where each of your inputs is placed. I strongly advise removing the continuum fitting line in the config file.


2. Then run line 27 
        
        config_fname = spec_path+'/config_AtomState.dat'
   
   "AtomState" must be replaced with whatever atom and state were given when you wrote the config file. The code will take all the transitions for the given AtomState in the given wavelength region/s and fit for 1 set of parameters for all transitions of the same AtomState.
   
3. Line 29 is where the fitting is done. **Be sure your config file is correct before you run this line.** 
    
        setup.bvp_mcmc(config_fname)
        
    This may take a few minutes to run. 

4. If your config file has the line "! auto 2" or something similar, you can go ahead and run the remainder of the code as is. 
    
    If you have given the "Maximum number of components to try" as 1, you will not have this line. You have to comment lines 34 and 45-53 and run the rest.
    
    You will now have a directory "bvp_output" in your spectrum_path. You will find all outputs here.
    
5. Another example of a config file is given as follows:
    
       1  spec_path /path/to/your/spectrum
       2  output o6
       3  mcmc 100 10000 8 bic kombine
       4  %% pg0003.spec 1387.0 1392.0 1397.0 1400.0
       5  % O VI 14 20 0.3479
       6  % O VI 16 20 0.3476
       7  logN 12.0 16.0
       8  b    0.0 40.0
       9  z    0.34775 100.0 
       10 lsf 1390 1400
    
    Any HST/COS spectrum can be used. the file format must be ascii with no header lines and 3 columns - wavelength, normalized flux and flux error in this same order. 
    
    Line 4 can have as many spectrum ranges as available. One just has to make sure there is no LSF or the number of LSF files matches the number of spectrum intervals
    
    Lines 5 and 6 are used to add components. If there are more components, one can simply add them in the same format. Atleast 1 component is necessary to run the fit. One can otherwise increase the 
    
    logN, b and z are the prior ranges and can be changed as per requirement.
    
    Lines 31 onwards is only the plotting part... you can change the plotting lines however you wish. 
    
    To change plotting part of the code, Ctrl+Click on the plotting functions like "plot_model_comparison" or "corner_plot"
